```python
# Render our plots inline
#%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
from random import randint

def flipper(trials):
    coin_list = []
    dice_list  = []
    for each_flip in range(trials):
        coin = randint(0,100)
        if coin <= 50:
            coin_list.append('Heads')
        if coin > 50:
            coin_list.append('Tails')
        die = randint(1,6)
        dice_list.append(die)
        
    return coin_list, dice_list


times_to_flip = 100000
coins, dice = flipper(times_to_flip)
mydf = pd.DataFrame([coins, dice])
mydf = mydf.T
mydf.columns = ['Coins', 'Dice']
mydf['Ints'] = np.where(mydf['Coins'] == "Heads", 1, -1)
mydf['CumSum'] = mydf.Ints.cumsum()
mydf['CumSum'].plot()
```

```python
times_to_flip = 10000
tests = 10
score = 0
score_list = []

for each in range(tests):
    coins, dice = flipper(times_to_flip)
    mydf = pd.DataFrame([coins, dice])
    mydf = mydf.T
    mydf.columns = ['Coins', 'Dice']
    mydf['Ints'] = np.where(mydf['Coins'] == "Heads", 1, -1)
    mydf['CumSum'] = mydf.Ints.cumsum()
    mydf['CumSum'].plot()
    score = mydf['Ints'].sum()
    score_list.append(score)

score_list = np.array(score_list)
score_list.sum()
```

```python
import math
mydf["Consecutives"] = 0
inarow = 1
for i in range(len(mydf)):
    if i+1 >= len(mydf): break
    current = mydf["Coins"][i]
    current_plus_one = mydf["Coins"][i+1]
    if current_plus_one == current:
        inarow += 1
    else: inarow = 1
    mydf["Consecutives"][i+1] = inarow


mydf['the_chances'] = 0.5 ** mydf['Consecutives'] 
# mydf.sort("Consecutives", ascending=False)
mydf
```
