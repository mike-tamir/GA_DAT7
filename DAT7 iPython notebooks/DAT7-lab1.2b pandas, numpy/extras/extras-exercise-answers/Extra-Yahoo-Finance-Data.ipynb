```python
import pandas as pd
import pandas.io.data

df = pd.read_csv("../exercise-data/hospital-costs.csv")
colz = ['Mean Charge', 'Median Charge', 'Mean Cost', 'Median Cost']
for each in colz:
    df[each] = df[each].str.replace('$', '')
    df[each] = df[each].astype(float)
df.to_csv('../exercise-data/hospital-costs-clean.csv')
df['Total Charge'] = df['Discharges'] * df['Mean Charge']
df['Total Cost'] = df['Discharges'] * df['Mean Cost']
df['Markup'] = df['Mean Charge'] / df['Mean Cost']
# df.head()
```
