```python
from random import randint
print randint(0,1)
def flipper(trials):
    coin_list = []
    dice_list = []
    for each_flip in range(trials):
        coin = randint(0,1)
        if coin == 1:
            coin_list.append('Heads')
        if coin == 0:
            coin_list.append('Tails')
        die = randint(1,6)
        dice_list.append(die)
        
    return coin_list, dice_list

times_to_flip = 100000
coins, dice = flipper(times_to_flip)
```

```python
mydf = pd.DataFrame(coins)
mydf.columns = ['Coins']
```

```python
# mydf['Coins'] == Heads fill new column with 1
# else fill new colum with -1
```

```python
#np.where(condition, fill with this, else fill with this)
mydf['Payout'] = np.where(mydf['Coins'] == "Heads", 1, -1)
```

```python
mydf['CumSum'] = mydf['Payout'].cumsum()
```
