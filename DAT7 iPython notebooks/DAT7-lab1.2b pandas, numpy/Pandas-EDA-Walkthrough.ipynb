```python
# THIS IS WHY THEY CALL DataFrames 'JU MAX' IN FRENCH
# CAUSE THEY ARE THE BEST
# Rename Columns
# READ IN A DATA FILE

df = pd.read_csv('lecture-data/city-of-chicago-salaries.csv')
df.columns = ['Name', 'Job', 'Department', 'Salary']

grouped = df.groupby('Department')
ppl_counts = grouped['Job'].value_counts()
print ppl_counts
avg_salary = grouped['Salary'].mean()

newdf = pd.DataFrame([ppl_counts, avg_salary])
newdf = newdf.T

newdf.columns = ['PeopleCount', 'AvgSalary']
newdf.plot(kind='bar')
```

```python
import numpy as np
import random
randomness = []
for i in xrange(100000):
    x = random.random()
    xs = ("{0:.2f}".format(round(x,1)))
    randomness.append(xs)

rdf = pd.DataFrame([randomness])
rdf = rdf.T
rdf.columns = ['Numbers']
```

```python
rdf['Numbers'].value_counts().plot(kind='bar')
```
