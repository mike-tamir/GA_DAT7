```python
# Rename Columns
df.columns = ['Name', 'Job', 'Department', 'Salary']
df.head()
```

```python
# Create a new column
df['My New Column'] = "F the Police"
df['AnotherNewColumn'] = df['Salary'] * -2
```

```python
# Delete or 'drop' a column
df = df.drop("My New Column", axis=1)
df = df.drop("AnotherNewColumn", axis=1)
```

```python
# Duplicate a column
df['NameDupe'] = df['Name']

# Drop that column cause its stupid
df = df.drop('NameDupe', axis=1)
```

```python
# Sort by a column
df.sort("Salary",ascending=False)

# Sort by multiple columns
df.sort(["Salary","Job"],ascending=True)
```
